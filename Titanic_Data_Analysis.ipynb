# Titanic Survival Rate Analysis

The tragedy of Titanic saddens everyone. From the movie, we have noticed that women and children had priorities to get on the survival boats, also few amount of higher class people managed to survive. In the analysis, we would like to dig into the passenger information dataset (provided by **Kaggle**) and find out the truth. 

The question we are trying to answer in this analysis:
**Is there a gender gap? Did female have higher survival rate than males?** 

The [dataset from Kaggle](https://www.kaggle.com/c/titanic): 
**Titanic Data** - Contains demographics and passenger information from 891 of the 2224 passengers and crew on board the Titanic. 


## Load Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
from scipy.stats import chisquare
import numpy as np
titanic_df = pd.read_csv("titanic-data.csv")
titanic_df.shape
titanic_df.head()
titanic_df.info()
```

## Data cleaning

From the description of the loaded dataframe, we can see there is missing data in columns 'Age', 'Cabin' AND 'Embarked'.
We will fill the gaps using  `fillna(method='pad')` function, so we could preserve variable's distribution.

```python
# Fill missing age

#fillna(method='pad') fill gaps while preserving variable's distribution
titanic_df['Age'] = titanic_df['Age'].fillna(method='pad')

# Fill missing class

titanic_df['Embarked'] = titanic_df['Embarked'].fillna(method='pad')

# Add new column to catergorize the age_group

titanic_df['Age_Group'] = float('NaN')
titanic_df.loc[titanic_df['Age'] >= 18, 'Age_Group'] = 'Adult'
titanic_df.loc[titanic_df['Age'] < 18 , 'Age_Group'] = 'Child'
titanic_df.loc[titanic_df['Age'] <= 6, 'Age_Group'] = 'Enfant'  

# Create Survival Label Column
titanic_df['Survival'] = titanic_df.Survived.map({0 : 'Dead', 1: 'Survived'})
titanic_df.Survival.head()
# Create Pclass Label Column
titanic_df['Class'] = titanic_df.Pclass.map({1 : 'First Class', 2 : 'Second Class', 3 : 'Third Class'})
titanic_df.Class.head()
# Create Embarked Labels Column
titanic_df['Ports'] = titanic_df.Embarked.map({'C' : 'Cherbourg', 'Q' : 'Queenstown', 'S' : 'Southampton'})
titanic_df.Ports.head()
```

This bar plot shows the survival rate of female vs male. Only 18.9% of male survived whereas 74.2% of female survived. 
From the visualization, we have our **initial hypothesis**: 
>Females were more likely to survive in this incident. 

We would like to perform a statistic test to verify our hypothesis. We will use Chi-square test in this case.

### Chi-Squared Test
#### Problem
>Is there a gender gap? Did female have higher survival rate than males?

**Chi-Squared Equation:**
$$ {\chi}^2=\frac{1}{d}\sum_{k=1}^{n} \frac{(O_k - E_k)^2}{E_k}\ $$

#### Solution

- **Hypotheses Statement**
> - H<sub>0</sub>: Survival rate of a passenger is independent of his/her sex
> - H<sub>a</sub>: Survival rate of a passenger is not independent of his/her sex<br\>

Significance level of 0.05

- **Observed Data**

```python
observed = titanic_df.groupby(['Sex'])['Survival'].value_counts()
observed
```

```python
observed_value = scipy.array([233,81,468,109])
```

- **Expected Data**
> The expected value if H<sub>0</sub> is true, which means female and male would have equal probability of survival. We have calculated earlier, the sample population has a probablity of survival of 38.38%. 

```python
rate_survived
survived=num_survived[1]/2
survived
non_survived = num_survived[0]/2
non_survived
```

```python
expected_value = scipy.array([survived,non_survived,non_survived,survived])
```

#### Test Statistics & P value

```python
chisquare(observed_value, f_exp=expected_value)
```

>Test Statistics = 318.26
>p-value way less than 0.05


### Conclusion
Based on the test statistics and p-value of Chi-squared test, we reject the null Hypothesis and conclude that the survival rate is correlated with the sex of the passenger. 

We have got an p-value which is way smaller than the significance level of 0.05. Therefore, we have strong evidence that female passengers indeed have higher rate of survival than male passengers. 

However, as we have mentioned earlier, we have limitation in our analysis. We assumed that sex is independent of other features of the passenger. At the same time, we have a lot of missing data inputs. Thus, we're limited to see the population tendancy. Which means that, the sex feature may not be independent if we had all the age information. I.E if the missing age are mostly less than 10 for females, then instead of being correlated to Sex, we would suspect survival rate be correlated to Age.
